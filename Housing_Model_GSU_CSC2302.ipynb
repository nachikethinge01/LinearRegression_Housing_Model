```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot

from sklearn.linear_model import LinearRegression #Importing Modules
```

```python
chart = pandas.read_csv('Cleaned_data_01.csv')
```

```python
chart.describe()
```

```python
chart.head()
```

```python
a = chart[['Cost_Living']]
b = chart[['sqft_living']]
```

```python
matplotlib.pyplot.figure(figsize=(10,6))
matplotlib.pyplot.scatter(a, b, alpha=0.3)
matplotlib.pyplot.title('Cost of living vs SquareFootage')
matplotlib.pyplot.xlabel('Cost of living')
matplotlib.pyplot.ylabel('Squarefootage of living')
matplotlib.pyplot.ylim(0,15000)
matplotlib.pyplot.xlim(0,9000000)
matplotlib.pyplot.show()
```

```python
regression = LinearRegression()
regression.fit(a,b)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
matplotlib.pyplot.figure(figsize=(10,6))
matplotlib.pyplot.scatter(a, b, alpha=0.3)
matplotlib.pyplot.plot(a, regression.predict(a), color='black', linewidth=4)
matplotlib.pyplot.title('Cost of living vs SquareFootage')
matplotlib.pyplot.xlabel('Cost of living')
matplotlib.pyplot.ylabel('Squarefootage of living')
matplotlib.pyplot.ylim(0,15000)
matplotlib.pyplot.xlim(0,9000000)
matplotlib.pyplot.show()
```

```python
regression.score(a,b)
```
